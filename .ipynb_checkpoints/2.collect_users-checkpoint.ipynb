For some reason the verified user fiel does not have follower/following information (which is critical for our analysis). Unfortunately this must be scraped as well. We use current follower/following data, acknowleding that there may be differences from August 2018.

```python
import pandas as pd
import twitterscraper
from bs4 import BeautifulSoup
import csv
import datetime
import pickle
import os
import numpy as np
import sys
import time
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from twitterscraper import query_user_info
from twitterscraper import query_tweets
from twitterscraper import query_tweets_from_user
from twitterscraper import user
import numpy as np
import twitterscraper
from tqdm import tqdm
import pickle

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
df = pd.read_pickle('verified_tweets.pkl')
df = pd.DataFrame(df)
df = df[df['timestamp'] < datetime.date(year=2019,month=1,day=1)]
verified_users = df['screen_name'].unique()
```

```python
existing_df = pd.read_pickle('user_accounts.pkl')
# df = df[df['timestamp'] < datetime.date(year=2019,month=1,day=1)]
run_users = existing_df['user.user'].unique()
```

```python
verified_users = set(verified_users) - set(run_users)
len(verified_users)
```

```python
total_users = len(verified_users)
```

```python
# with open('verified_users.txt', 'rb') as fp:
#     verified_users = pickle.load(fp)
```

```python
def user_info(user_pull):
    try:
        user = query_user_info(user_pull)
        if(user):
            return({
                'user.id' : user.id,
                'user.user' : user.user,
                'user.full_name' : user.full_name,
                'user.date_joined': user.date_joined,
                'user.followers' : user.followers,
                'user.following' : user.following
            })
        else:
            return({
                'user.id' : np.nan,
                'user.user' : user_pull,
                'user.full_name' : np.nan,
                'user.date_joined': np.nan,
                'user.followers' : np.nan,
                'user.following' : np.nan
            })
    except:
        return({
            'user.id' : np.nan,
            'user.user' : user_pull,
            'user.full_name' : np.nan,
            'user.date_joined': np.nan,
            'user.followers' : np.nan,
            'user.following' : np.nan
        })
```

```python
import logging
logger = logging.getLogger()
logger.disabled = True
userlist = []
for ix, user_pull in enumerate(verified_users):
    print(ix / total_users)
    userlist.append(user_info(user_pull))
    if ix % 20 == 0:
        df = pd.DataFrame(userlist)
        existing_df = existing_df.append(df, ignore_index=True)
        existing_df.to_pickle('user_accounts.pkl')
        userlist = []

logger.disabled = False
```
