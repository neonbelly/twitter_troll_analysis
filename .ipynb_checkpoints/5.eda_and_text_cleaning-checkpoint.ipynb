Things to note:
There is a 5,000 friends limit. That is if you have less than 5,000 followers, the maximum you can follow is 5,000 accounts.
The 10% limitâ€ . If you follow more than 5,000 users, you'll be limited to 10% more than the number of people that follow you. For example, if 5,000 people follow you, you can follow a max of 5,500 people, if 10,000 people follow you, you can follow up to 11,000 people, and so on.

# Text Analysis

```python
df.head()
```

Clean text for analysis

```python
def text_clean(text):
    text = text.lower()
    text = re.sub(r'http\S+', '', text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
    text = text.strip(' ')
    text = text.strip('\n')
    text = re.sub('[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub('_', '', text)
    return text
```

```python
df['text']=df['text'].astype(str)
df['text_cleaned'] = df['text'].map(lambda tweet: text_clean(tweet))
```

Engineer features from text

```python
df['upper_count'] = df['text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df['char_count'] = df['text'].str.len()
df['word_count'] = df['text_cleaned'].apply(lambda x: len(str(x).split(" ")))
```

```python
df['hashtags'] = df['hashtags'].str.strip('][')
df['hashtags'] = df['hashtags'].apply(lambda l: ''.join([str(x).strip('#') for x in l]))
df['hashtag_count'] = df['hashtags'].apply(lambda x: len(str(x).split(",")) if x != '' else 0)
```
